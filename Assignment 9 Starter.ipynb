#### Task 1: Import libraries

#### Task 2: Open file
Open the file called 'tripadvisor_review.csv' and display the first 10 rows.

```python
file_name = 'tripadvisor_review.csv'
df = pd.read_csv(file_name)
df.head(10)
```

#### Task 3: Fit 2-cluster model

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2, random_state=42)
df['cluster'] = kmeans.fit_predict(df)

print("Cluster Centers:\n", kmeans.cluster_centers_)
```

#### Task 4: Print Centroid Values
Centroid Values for 2-cluster model

```python
for i, center in enumerate(kmeans.cluster_centers_):
    print(f"Centroid {i + 1}: {center}")
```

#### Task 6: Identify the number of observations in each cluster

Expected output:

```python
cluster_counts = df['cluster'].value_counts().sort_index()

for cluster_label, count in cluster_counts.items():
    print(f"cluster {cluster_label}: {count}")
```

#### Task 7: Calculate Silhouette Score

```python
from sklearn.metrics import silhouette_score

features = df.drop('cluster', axis=1)

score = silhouette_score(features, df['cluster'])

print(f"Silhouette Score: {score:.4f}")
```

#### Task 8: Repeat for a 3-cluster model
Generate a new model using 3 clusters instead of 2

Repeat all of the steps from previous tasks to generate the model and print centroid values, labels, number of observations, and silhouette score.

We will compare the centroid values and silhouette scores to determine which model is a better fit for the data


```python
from sklearn.cluster import KMeans

# Fit KMeans with 3 clusters
kmeans_3 = KMeans(n_clusters=3, random_state=42)
df['cluster_3'] = kmeans_3.fit_predict(df.drop(['cluster'], axis=1))

print("Centroid Values for 3-cluster model:")
for i, center in enumerate(kmeans_3.cluster_centers_):
    print(f"Centroid {i + 1}: {center}")

print("Cluster labels for a subset of the observations:")
print(df['cluster_3'].values[:20])

cluster_counts_3 = df['cluster_3'].value_counts().sort_index()
for label, count in cluster_counts_3.items():
    print(f"cluster {label}: {count}")

from sklearn.metrics import silhouette_score

# Calculate silhouette score (use only features, exclude both label columns)
features = df.drop(['cluster', 'cluster_3'], axis=1)
sil_score_3 = silhouette_score(features, df['cluster_3'])

print(f"Silhouette Score for 3-cluster model: {sil_score_3:.4f}")
```
