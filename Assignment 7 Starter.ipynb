#### Task 1

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels. formula.api as smf
```

#### Task 2
- Open the file _usedcars.csv_ using Pandas and store in a dataframe variable called _cars_
- Print the first 10 rows of the dataframe to the notebook

```python
cars = pd.read_csv('usedcars.csv')
cars.head(10)
```

#### Task 3
- Define the regression model and store the result in a variable called _model_.  The regression model should predict price by using year, mileage, and transmission_code as predictors. You should use the _cars_ dataframe as the data source for this analysis.
- Fit the model and store the result in a variable called _result_.

```python

model = smf.ols('price ~ year + mileage + transmission_code', data=cars)
result = model.fit()
```

#### Task 4
- Print the result summary

```python

print(result.summary())
```
