```python
%%file MovieRatingAverage.py
from mrjob.job import MRJob
from mrjob.step import MRStep
import statistics

class MovieRatingAverage(MRJob):
    def steps(self):
        return [
            MRStep(mapper=self.mapper_get_ratings,
                   reducer=self.reducer_avg_ratings)
        ]

    def mapper_get_ratings(self, _, line):
        userID, movieID, rating, timestamp = line.split('\t')
        yield int(movieID), int(rating)

    def reducer_avg_ratings(self, movieID, ratings):
        ratings_list = list(ratings)
        yield movieID, statistics.mean(ratings_list)

if __name__ == "__main__":
    MovieRatingAverage.run()
```
