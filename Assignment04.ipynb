```python
import sys

def RPSValidate(player_input):
    valid = ['rock', 'paper', 'scissors']
    if player_input.lower() in valid:
        return player_input.lower()
    return 'input error'

def RPSExecute(player1, player2):
    if player1 == player2:
        return 0
    elif (player1 == 'rock' and player2 == 'scissors') \
         or (player1 == 'paper' and player2 == 'rock') \
         or (player1 == 'scissors' and player2 == 'paper'):
        return 1
    else:
        return 2

attempts = 3
while attempts > 0:
    player1 = input('Enter value for Player 1: ')
    player1 = RPSValidate(player1)
    if player1 != 'input error':
        break
    print('Invalid input')
    attempts -= 1
else:
    print('Player 1 exceeded the allowable number of tries.')
    sys.exit()

attempts = 3
while attempts > 0:
    player2 = input('Enter value for Player 2: ')
    player2 = RPSValidate(player2)
    if player2 != 'input error':
        break
    print('Invalid input')
    attempts -= 1
else:
    print('Player 2 exceeded the allowable number of tries.')
    sys.exit()

result = RPSExecute(player1, player2)

if result == 0:
    print('Players Tie')
elif result == 1:
    print('Player 1 Wins')
else:
    print('Player 2 Wins')

```
